```python
#bt áp dụng 1
# Kho hàng ban đầu
cua_hang = {"do_noi_that": ["ghe", "ban", "sofa"], "so_luong": [24, 7, 6], "gia": [200, 500, 1200]}
print("Kho hàng ban đầu:", cua_hang)

# a. Khách hàng mua 4 chiếc ghế
vi_tri_ghe = cua_hang["do_noi_that"].index("ghe")
cua_hang["so_luong"][vi_tri_ghe] -= 4
print("\nSau khi bán 4 ghế:", cua_hang)

# b. Thêm các món đồ nội thất mới sử dụng các cú pháp khác nhau
# Phương pháp 1: Sử dụng append
cua_hang["do_noi_that"].append("tham")
cua_hang["so_luong"].append(9)
cua_hang["gia"].append(150)

# Phương pháp 2: Sử dụng phép nối danh sách
cua_hang["do_noi_that"] += ["den"]
cua_hang["so_luong"] += [4]
cua_hang["gia"] += [180]
print("\nSau khi thêm các mặt hàng mới:", cua_hang)

# c. Chủ nhà hàng mua tất cả các bàn - sử dụng hai phương pháp khác nhau
# Phương pháp 1: Gán trực tiếp
vi_tri_ban = cua_hang["do_noi_that"].index("ban")
cua_hang["so_luong"][vi_tri_ban] = 0

# Phương pháp 2: Cách tiếp cận thay thế (nếu chúng ta cần hiển thị phương pháp khác)
# Chúng ta có thể sử dụng: cua_hang["so_luong"][vi_tri_ban] -= cua_hang["so_luong"][vi_tri_ban]
print("\nSau khi bán tất cả các bàn:", cua_hang)

# d. In từ điển với các khóa căn phải và các giá trị căn trái
print("\nKho hàng hiện tại:")
for khoa in cua_hang:
    print("{:>10}: {}".format(khoa, cua_hang[khoa]))

# e. Tính tổng giá trị kho hàng
tong_gia_tri = 0
for i in range(len(cua_hang["do_noi_that"])):
    tong_gia_tri += cua_hang["so_luong"][i] * cua_hang["gia"][i]
print("\nTổng giá trị kho hàng: ${}".format(tong_gia_tri))
```

```python
#bt2
# Từ điển ban đầu
tu_dien = {"numbers": [2, 3, 4, 5, 6, 7, 8, 9, 10]}
print("Từ điển ban đầu:", tu_dien)

# a. Thêm cặp khóa:giá trị với khóa "even" và giá trị là danh sách Boolean
tu_dien["even"] = []
for so in tu_dien["numbers"]:
    if so % 2 == 0:  # Kiểm tra số chẵn
        tu_dien["even"].append(True)
    else:
        tu_dien["even"].append(False)
print("\nSau khi thêm danh sách Boolean:", tu_dien)
print("Danh sách Boolean:", tu_dien["even"])

# b. Trừ 1 cho mỗi số
for i in range(len(tu_dien["numbers"])):
    tu_dien["numbers"][i] -= 1
print("\nSau khi trừ 1 cho mỗi số:", tu_dien)
print("Danh sách số mới:", tu_dien["numbers"])
print("Danh sách Boolean (chưa cập nhật):", tu_dien["even"])

# c. Sửa đổi danh sách Boolean bằng cách dịch chuyển
# Lưu giá trị cuối cùng
gia_tri_cuoi = tu_dien["even"][-1]
# Dịch chuyển các giá trị sang phải
for i in range(len(tu_dien["even"]) - 1, 0, -1):
    tu_dien["even"][i] = tu_dien["even"][i-1]
# Gán giá trị cuối cùng cho vị trí đầu tiên
tu_dien["even"][0] = gia_tri_cuoi

print("\nSau khi dịch chuyển danh sách Boolean:", tu_dien)
print("Danh sách số:", tu_dien["numbers"])
print("Danh sách Boolean (đã cập nhật):", tu_dien["even"])

# Kiểm tra xem danh sách Boolean có tương ứng với danh sách số mới không
print("\nKiểm tra tính đúng đắn:")
for i in range(len(tu_dien["numbers"])):
    so = tu_dien["numbers"][i]
    la_chan = tu_dien["even"][i]
    print("Số {} là chẵn: {} - Đúng không? {}".format(so, la_chan, (so % 2 == 0) == la_chan))
```

```python
#bt 3
def tao_tam_giac(n):
    """Tạo từ điển chứa hình tam giác số với n hàng."""
    tam_giac = {}
    for hang in range(1, n + 1):
        tam_giac[hang] = []
        for so in range(1, hang + 1):
            # Thêm số vào danh sách, lặp lại số lần bằng chính giá trị của số
            tam_giac[hang].extend([so] * so)
    return tam_giac

def in_tam_giac(tam_giac):
    """In hình tam giác từ từ điển."""
    for hang, danh_sach in tam_giac.items():
        print("Hàng {}: {}".format(hang, danh_sach))

def choi_game():
    """Cho phép người chơi tạo các hình tam giác số."""
    print("Chào mừng đến với trò chơi Tam Giác Số!")
    
    while True:
        try:
            n = int(input("\nNhập một số nguyên dương (hoặc 0 để thoát): "))
            if n == 0:
                print("Cảm ơn bạn đã chơi! Tạm biệt!")
                break
            elif n < 0:
                print("Vui lòng nhập một số nguyên dương.")
                continue
                
            # Tạo và in tam giác
            tam_giac = tao_tam_giac(n)
            print("\nHình tam giác với {} hàng:".format(n))
            in_tam_giac(tam_giac)
            
        except ValueError:
            print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")

# Bắt đầu trò chơi
choi_game()
```

```python
#bai 2
def multiplication_table_game():
    print(" CHÀO MỪNG ĐẾN VỚI TRÒ CHƠI BẢNG CỬU CHƯƠNG! ")
    print("Nhập một số và xem bảng cửu chương 1-10")
    print("Nhập 'thoát' khi bạn muốn kết thúc trò chơi.")
    
    while True:
        # Nhận đầu vào từ người dùng
        user_input = input("Nhập một số (hoặc '0' để kết thúc): ")
        
        # Kiểm tra nếu người dùng muốn thoát
        if user_input.lower() in ['0']:
            print("Cảm ơn bạn đã chơi! Hẹn gặp lại! ")
            break
        
        try:
            # Chuyển đổi đầu vào thành số nguyên
            number = int(user_input)
            
            # Tạo từ điển bảng cửu chương
            multiplication_dict = {i: i * number for i in range(1, 11)}
            
            # Hiển thị bảng cửu chương
            print(" BẢNG CỬU CHƯƠNG CỦA", number )
            for key, value in multiplication_dict.items():
                print(number, "x", key, "=", value)
                
            # Thêm một chút khích lệ
            if number <= 10:
                print("Tuyệt vời!  Bạn đang làm rất tốt!")
            else:
                print("Wow! Bạn đang học những số lớn! ")
                
        except ValueError:
            print("Oops! Đó không phải là một số hợp lệ. Hãy thử lại! ")

# Chạy trò chơi
if __name__ == "__main__":
    multiplication_table_game()
```
